#### Dataframe: listing each room_id, host_id with total score in two sorting ways

1) index = (room_id, host_id) <br>
2) column = total_score: overall_satisfaction + reviews * 0.378 <br>
3) output = 1.sorted total_score in ascending 2.sorted total_socre in descending
            = sorted_total_score_ascend.csv, sorted_total_score_descend.csv

```python
import pandas as pd
import numpy as np

#Total data
Total_data = []
# open the csv file
with open('./airbnb.csv','r') as fileRead:
    i = 0
    # read the data by the line
    while i < 3279:
        # read the data in one line
        lineContent = fileRead.readline()
        # append the data to if the line is not empty and \n
        if(lineContent != '\n') and (lineContent != ''):
            Total_data.append(lineContent.strip('\n').split(','))
        i+=1

# create the array to store the data
List_data = []
len_total_data = len(Total_data)

for i in range(len_total_data):
    # create the middle array
    List_middle_data = []
    # except the first row
    if i== 0:
        #List_middle_data.append(Total_data[0][0])
        #List_middle_data.append(Total_data[0][1])
        #List_middle_data.append('total score')
        #List_data.append(List_middle_data)
        i += 1
        continue

    # calculate the total score
    total_score = 0.0
    List_middle_data.append(Total_data[i][0])
    List_middle_data.append(Total_data[i][1])
    # check the special case
    if(Total_data[i][6] == ''):
        Total_data[i][6] = 0
    total_score += float(Total_data[i][6]) + float(Total_data[i][5]) * 0.378
    List_middle_data.append(total_score)
    List_data.append(List_middle_data)


# use the dataframe type
# Total_data
df_origin = pd.DataFrame(Total_data)
# extract data
df_aim = pd.DataFrame(List_data)
```

## Sort & save as files

```python
import csv

# sort by ascending order
ascending_data = List_data.sort(key=lambda x: x[2])
pd_ascending = pd.DataFrame(List_data)
print('***[Ascending Order]***')
print(pd_ascending)
output1 = open("./sorted_total_score_ascend.csv", 'w')
myWriter1 = csv.writer(output1, delimiter = ",")
myWriter1.writerow(['room_id', 'host_id', 'total_score(ascending)'])
myWriter1.writerows(List_data)
output1.close()

# sort by descendig order
descending_data = List_data.sort(key=lambda x: x[2], reverse=True)
pd_descending = pd.DataFrame(List_data)
print('\n***[Descending Order]***')
print(pd_descending)
output2 = open("./sorted_total_score_descend.csv", 'w')
myWriter2 = csv.writer(output2, delimiter = ",")
myWriter2.writerow(['room_id','host_id', 'total_score(descending)'])
myWriter2.writerows(List_data)
output2.close()
```

```python
import pandas as pd
import numpy as np


file_name = 'D:\\jupyter\\19_2Bigdata\\airbnb.csv'
df_air = pd.read_csv(file_name)      # read the file
#df_air
df_air1 = df_air.set_index(['neighborhood'])  # set neighborhodd as an index
neigh_list = set(df_air1.index.tolist()) # index를 lis로 가져오기
#print(neigh_list)

# make a empty dataframe
df_result = pd.DataFrame(index = neigh_list)

total_avg_reviews = []  # 전체 각 category 별 저장
total_avg_overall_satisfaction = []
total_max_price = []


# process the data
for neighbor in neigh_list:
    df_air2 = df_air1[df_air1.index == neighbor] # extract the portion
    newdf = df_air2.copy()   # copy
    avg_reviews= newdf['reviews'].mean()
    avg_of_overall_satisfaction= newdf['overall_satisfaction'].mean()
    avg_of_price = newdf['price'].mean()
    max_of_reviews = newdf['reviews'].max()
    min_of_reviews = newdf['reviews'].min()
    max_of_price = newdf['price'].max()
    min_of_price = newdf['price'].min()
    total_avg_reviews.append(avg_reviews)
    total_avg_overall_satisfaction.append(avg_of_overall_satisfaction)
    total_max_price.append(max_of_price)
    
df_result['avg reviews'] = total_avg_reviews
df_result['max price'] = total_max_price

df_result

```
